#Adicionando Linhas

Existe um documento contendo as vendas de dezembro, para adicionalo no final usando:

 pd.concat([onde vai ser adicionado, o que vai ser adicionado], ignore_index = True)

###cuidado, se for executado repetidas vezes, repetidas e redundantes linhas serão adicionadas###

```python
vendas_dez = pd.read_excel("/content/drive/MyDrive/04-12 - Introdução ao Pandas/Vendas - Dez.xlsx")
#display(vendas_dez)
tabela_vendas = pd.concat([tabela_vendas,vendas_dez], ignore_index=True)
display(tabela_vendas)
```

# Adicionando coluna

- a partir de uma coluna pré-existente

- criando uma coluna nova com valores padrão

```python
# criar uma comissão de 5% para o vendedor em cima do valor final
tabela_vendas['Comissão'] = tabela_vendas['Valor Final'] * 0.05
display(tabela_vendas)

# criar uma coluna de imposto que ainda não tenho conhecimento dos valores
tabela_vendas.loc[:, 'Imposto']=0 # o':' significa que todos os valores serão preenchidos pelo valor presente depois do '='
display(tabela_vendas)
```

#Valores vazios

- deletar linhas e colunas vazias .dropna(how='x', axis ='y')
  - x = all -exclui linhas e colunas que estejam completamente vazias
  - por padrão ele considera linha, se informar y = 1 ele passa a considerar coluna

- deletar linhas e colunas que contenham pelo menos um valor vazio
 - .dropna() - sem nenhum argumento dentro dos parenteses

- preencher valores vazios (média e último valor)

- .fillna(argumentos)
- .ffill()

```python
tabela_vendas = tabela_vendas.dropna(how='all') # exclui todas as linhas em que todos os valores são vazios
tabela_vendas = tabela_vendas.dropna(how='all', axis=1) # exclui todas as colunas em que todos os valores são vazios

tabelas_vendas = tabela_vendas.dropna() # exclui linhas que possuem pelo menos um valor vazio

tabela_vendas['Comissão'] = tabela_vendas['Comissão'].fillna(tabela_vendas['Comissão'].mean()) # Atribui a todos os valores nan de comissão a média dos valores de comissão (media realizada pelo comando .mean())
display(tabela_vendas) #como não havia valor vazio, não houve alterção

#Para tabelas "incompletas" em excel
tabela_vendas = tabela_vendas.ffill() # preenche linhas que estão vazias com o valor de cima
```

#Calculando indicadores

- Groupby
 - .groupby('x').y agrupa todos os elementos de x e realiza o calculo y com as outas colunas

- Value Counts

 - ['x'].value_counts() conta quantas coisas existem dentro de x


```python
n_transacao = tabela_vendas['ID Loja'].value_counts() # conta quantos vendas foram realizadas em cada loja
# display(n_transacao)

faturamento_produto = tabela_vendas[['Produto', 'Quantidade', 'Valor Final']].groupby('Produto').sum() # mostra uma nova tabela ordenada por produtos vendidos, exibindo apenas oq foi ordenado entre [[]], o .sum() indica soma, se eu quisesse media era só substituir por .mean()
display(faturamento_produto)
```
